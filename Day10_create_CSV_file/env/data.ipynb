```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
df = df[['Sales_Amount' , 'Quantity_Sold' ]]
```

```python
df = df.rename(columns={'Sales_Amount' : 'Amount' , 'Quantity_sold' : 'Quantity'})
```

```python
df.head()
```

```python
df.info()
```

```python
import random
df['Discount'] = random.choices([0,5,10,15,20,25,30,35,40,45,50],weights=[10,10,10,10,10,1,1,1,1,10,10],k=1000)
```

```python
df['Order_id']=range(1001,2001)
```

```python
df.info()
```

```python
df
```

```python
df = df.loc[:,['Order_id' , 'Amount' , 'Discount' , 'Quantity_Sold']]
```

```python
df.head()
```

```python
df.info()
```

```python
rows=10000

new_df = pd.DataFrame({
    "Order_ID": range(2001, 2001 + rows),
    "Amount": random.choices(range(1000, 10000), k=rows),
    "Discount": random.choices(
        [0,5,10,15,20,25,30,35,40,45,50],
        weights=[10,10,10,10,10,1,1,1,1,10,10],
        k=rows
    ),
    "Quantity_Sold": random.choices(range(1, 100), k=rows)
})
```

```python
new_df.tail()
```

```python
df.rename(columns={"Order_id": "Order_ID"}, inplace=True)
```

```python
df = pd.concat([df,new_df],ignore_index = True)
```

```python
df
```

```python
df.info()
```

```python
df.to_csv('sales.csv',index=False)
```
